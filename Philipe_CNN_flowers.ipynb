```python
# Instale o pacote do Kaggle para baixar o dataset
!pip install kagglehub

# Baixar o dataset de flores usando kagglehub
import kagglehub
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Baixar o dataset e confirmar o caminho
path = kagglehub.dataset_download("rahmasleam/flowers-dataset")
print("Path to dataset files:", path)

# Confirmar o conteúdo do diretório
print("\nConteúdo do diretório baixado:")
for root, dirs, files in os.walk(path):
    print(root, "contém subdiretórios:", dirs, "e arquivos:", files)

# Configurar o caminho correto para o diretório de imagens
train_dir = os.path.join(path, 'flower_photos')  # Ajustar o nome do subdiretório conforme listado acima

# Verificar se o diretório existe
if not os.path.exists(train_dir):
    raise FileNotFoundError(f"O diretório {train_dir} não foi encontrado. Verifique o caminho.")

# Configurações do dataset
img_height, img_width = 128, 128
batch_size = 32

# Gerador de dados para normalização e divisão de treino/validação
train_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

# Geradores de dados de treinamento e validação
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='sparse',  # para usar SparseCategoricalCrossentropy
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='sparse',
    subset='validation'
)

# Construção do modelo de classificação de flores
model = Sequential([
    Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # 5 classes para as diferentes espécies de flores
])

# Compilação do modelo
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Sumário do modelo
model.summary()

# Treinamento do modelo
epochs = 10
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs
)

# Avaliação do modelo e previsão
test_loss, test_acc = model.evaluate(validation_generator)
print(f"Test accuracy: {test_acc}")

# Previsões e matriz de confusão
pred = model.predict(validation_generator)
pred_classes = np.argmax(pred, axis=-1)
true_classes = validation_generator.classes
class_labels = list(validation_generator.class_indices.keys())

# Matriz de confusão
cm = confusion_matrix(true_classes, pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
